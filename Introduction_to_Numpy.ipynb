### **Numpy is faster than list**

```python
list1 = [i for i in range(100000)]
list2 = [i for i in range(100000)]

start = time.time()

list_tot = list1 + list2

end = time.time()

print(f"Runtime of the program is {round(end - start,5)}")
```

```python
np1 = np.array([i for i in range(100000)])
np2 = np.array([i for i in range(100000)])

start = time.time()

num_con = np.concatenate((np1, np2), axis=0)

end = time.time()

print(f"Runtime of the program is {round(end - start,5)}")
```

### **Join & Split**

#### Join

```python
# concatenate
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,11,12,13,14])

concat1 = np.concatenate((arr1,arr2), axis=0)
concat1
```

```python
# concatenate
arr3 = np.array([[1,2,3],[4,5,6]])
arr4 = np.array([[10,11,12],[13,14,15]])

concat2 = np.concatenate((arr3,arr4), axis=1)
concat2
```

#### Split

```python
# split
split_1 = np.array_split(concat1,2,axis=0)

display(split_1)

print("First array:", split_1[0])
print("Second array:", split_1[1]) 
```

```python
# split
split_2 = np.array_split(concat2,2,axis=1)

display(split_2)

print("First array:")
print(split_2[0])

print("Second array:")
print(split_2[1])
```

### **Sorting**

```python
# Descending 1D
array_sort = np.array([ 1,  2,  3,  4,  5, 10, 11, 12, 13, 14])
np.sort(array_sort)[::-1]
```

```python
# Descending 2D
array_sort = np.array([[10,11,12],[13,14,15]])
np.sort(array_sort)[::1, ::-1]
```

[::1] set for ascending the number of array, and [::-1] set for descending the element

### **Filter**

```python
arr = np.arange(1,20,2)
arr_ = np.arange(0,20,1)
display(arr)
display(arr_)
```

```python
display(np.where(arr < 5)) # index array

display(arr[np.where(arr < 5)]) # retrieve the value in array
```

```python
logical = np.where(arr < 5, "Kurang dari 5", "Lebih dari 5")

print(logical)
```

```python
arr_[np.where(arr_ % 2 == 0)] # return even element
```

### **Matrix Operation**

image.png

```python
# 1D Matrix operation
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,11,12,13,14])

np.add(arr1,arr2)
```

```python
# 2D Matrix operation
arr3 = np.array([[1,2,3],[4,5,6]])
arr4 = np.array([[10,11,12],[13,14,15]])

np.add(arr3,arr4)
```

```python
# 3D Matrix operation
arr5 = np.array([[[1,2,3],[4,5,6]], [[10,11,12],[13,14,15]]])
arr6 = np.array([[[12,13,14],[20,21,22]], [[30,31,33],[1,2,4]]])

np.add(arr5,arr6)
```
