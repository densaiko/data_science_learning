```python
# import library
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset
data = pd.read_csv('https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/Transformer.csv')
data.head()
```

## Scatterplot and Regplot
- **Scatter plot** is used to visualize two numerical variables and its relationship
- **Regplot** is similar with the scatterplot but it can emphasize the pattern and plot resulting with the 95% confidence level in default


```python
# Scatterplot
sns.scatterplot(data=data, x='CO', y='CO2')
plt.show()
```

```python
# Regplot
sns.regplot(data=data, x='CO', y='CO2')
plt.show()
```

```python
## Advanced Scatterplot
sns.scatterplot(data=data, x='CO', y='CO2', hue='event')
plt.show()
```

## Histplot and Displot
Histogram and Regplot have similar function to show the distribution of the data. But, Regplot is used more in figure-level


```python
# Histplot
sns.histplot(data=data, x='Nitrogen')
plt.show()
```

```python
# Histplot with probability
sns.histplot(data=data, x='Nitrogen', hue='event', stat='probability')
plt.show()
```

```python
# Histplot with CDF
sns.histplot(data=data, x='Nitrogen', hue='event', cumulative=True, stat='probability',
             element="step", fill=False)
plt.show()
```

```python
# Displot
sns.displot(data=data, x='Nitrogen', col='event')
plt.show()
```

```python
# Displot
sns.displot(data=data, x='Nitrogen', col='event', kind='ecdf')
plt.show()
```

## Catplot and Boxplot
- **Catplot** is used to visualize the scattered data of the numerical variable and categorical value.
- **Boxplot** is used not only to visualize the distribution of value but also to observe the extreme value (outlier)


```python
# Catplot
sns.catplot(data=data, x='event', y='Nitrogen')
plt.show()
```

```python
# Boxplot
sns.boxplot(data=data, x='event', y='Nitrogen')
plt.show()
```

```python
# Bonus countplot
sns.countplot(data=data, x='event')
plt.show()
```

## Pairplot

```python
data_2 = data.iloc[:, 13:18]
data_2.head()
```

```python
sns.pairplot(data_2, height=2.5)
plt.show()
```

```python
sns.pairplot(data_2, hue='event', height=2.5)
plt.show()
```
